```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Reading the file
df = pd.read_csv('dataset/Book1.csv')
```

```python
# Question 17
average_salary = np.mean(df["Salary"])
print(average_salary)
```

```python
# Question 18
filtered_df = df[df["Salary"] > 60000]
print(filtered_df)
```

```python
# Question 19
department_avg_salary = df.groupby('Department')['Salary'].mean()
department_avg_salary.plot(kind='bar', color='pink', width=0.4, edgecolor='black')
plt.title('Department-wise Average Salary')
plt.xlabel('Department')
plt.ylabel('Average Salary')
plt.show()

```

```python
df1 = pd.read_csv('dataset/Book2.csv')
df1
```

```python
# Question 20
department_employee_count = df1['Department'].value_counts()
department_employee_count.plot(kind='bar', color='lavender', width=0.3, edgecolor='black')
plt.title('Number of Employees in Each Department')
plt.xlabel('Department')
plt.ylabel('Number of Employees')
plt.show()
```

```python
# Question 21
filtered_df1 = df1[df1["Salary"] > 60000]
print(filtered_df1)
```

```python
# Question 22
salary_trend = df1.groupby('Department')['Salary'].mean()
salary_trend.plot(kind='line', x='Department', y='Salary', marker='o', color='blue')
plt.title('Salary Trend by Department')
plt.xlabel('Department')
plt.ylabel('Average Salary')
plt.grid()
plt.show()
```

```python
df2 = pd.read_csv('dataset/Book3.csv')
df2
```

```python
# Question 23
mean_scores = np.mean(df2[['Math', 'Science', 'English', 'History']], axis=0)
print(mean_scores)
```

```python
# Question 24
dataframe = pd.DataFrame(df2)
dataframe['Average Score'] = dataframe[['Math', 'Science', 'English', 'History']].mean(axis=1)
print(dataframe[['Student Name', 'Average Score']])
```

```python
# Question 25

dataframe = pd.DataFrame(df2)

# Adding a new column for total score
df["Total Score"] = dataframe.iloc[:, 1:].sum(axis=1)

# Finding the student with the highest overall score
top_student = dataframe.loc[df["Total Score"].idxmax()]

print(f"Student with the highest overall score:\n{top_student}")
```

```python

# Question 26
plt.scatter(df2['Math'], df2['Science'], color='orange', marker='o')
plt.title('Math vs Science Scores')
plt.xlabel('Math Scores')
plt.ylabel('Science Scores')
plt.grid()
plt.show()
```

```python
# Question 27
mean_scores = np.mean(df2[['Math', 'Science', 'English', 'History']], axis=0)
std_scores = np.std(df2[['Math', 'Science', 'English', 'History']], axis=0)

standardized_scores = (df2[['Math', 'Science', 'English', 'History']] - mean_scores) / std_scores
print(standardized_scores)
```

```python
# Question 29
subject_averages = dataframe[['Math', 'Science', 'English', 'History']].mean()
plt.figure(figsize=(8, 6))
plt.pie(subject_averages, labels=subject_averages.index,shadow = True, autopct='%1.1f%%', startangle=140)
plt.title('Subject-wise Average Scores')
plt.axis('equal') 
plt.show()
```

```python
# Question 30
average_history_score = dataframe['History'].mean()
students_above_average_history = dataframe[dataframe['History'] > average_history_score]
print("Students with History scores greater than the average:\n", students_above_average_history)
```

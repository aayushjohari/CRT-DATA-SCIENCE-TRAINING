```python
import seaborn as sns 
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# ignore warnings of ssl certificates

import ssl
ssl._create_default_https_context = ssl._create_unverified_context

# load dataset
data = sns.load_dataset("tips")
data.head()
```

```python
# Plot total_bill column
plt.plot(data.total_bill) 
plt.show()
```

```python
# Plot total_bill column
sns.lineplot(data["total_bill"]) 
```

```python
# Scatter plot of total_bill and tip columns
sns.scatterplot(x = 'total_bill', y = 'tip', data = data) 
plt.show()
```

```python
# Creates a histogram of given data
sns.histplot(data = data, x= "total_bill", color="pink")
plt.show()
```

```python
plt.hist(data["total_bill"], color="lavender", edgecolor="black")
plt.show()
```

```python
# Creates a Kernel Density Graph
sns.kdeplot(data["total_bill"], fill=True, color="violet") 
plt.show()
```

```python
# Scatter plot of total_bill and tip columns
sns.relplot(x = 'total_bill', y = 'tip', data = data) 
plt.show()
```

```python
data["time"].unique()
```

```python
# Creates a scatter plot with hue as time
sns.relplot(x = 'total_bill', y = 'tip', hue = 'time', data = data) 
plt.show()
```

```python
# Creates a scatter plot with time hue and tells which guys do smoking while doing that
sns.relplot(x = 'total_bill', y = 'tip', hue = 'time', style="smoker", data = data) 
plt.show()
```

```python
# Creates two scatter plots with tells who do smoking an dwho not in a separate graph
sns.relplot(x = 'total_bill', y = 'tip', hue = 'time', style="smoker", data = data, kind = "scatter", col = "smoker") 
plt.show()
```

```python
# Prints two graphs on the basis of time dependency
sns.relplot(x = 'total_bill', y = 'tip', data = data, kind = "scatter", hue = "sex", style = "smoker", col = "time") 
plt.show()
```

```python
data["day"].unique()
```

```python
# As their are four days it prints four graphs for each one
sns.relplot(x = "total_bill", y = "tip", data = data, hue = "time", col = "day", kind = "scatter", style = "smoker") 
plt.show()
```

```python
sns.relplot(x = "total_bill", y = "tip", data = data)
plt.title("Total Bill vs Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

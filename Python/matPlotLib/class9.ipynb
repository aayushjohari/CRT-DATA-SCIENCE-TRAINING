#                                                     **MATPLOTLIB**


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Plot, Scallater, Hist, Bar, Pie

# Created two points Arrays
xPoint = np.array([0,6])
yPoint = np.array([0,250])

# Used to plot the points on a graph
plt.plot(xPoint, yPoint) 

# Used to show the graph
plt.show()
```

```python
# Created two points Arrays
xPoint = np.array([1,8])
yPoint = np.array([3,10])

# Used to create a plot
plt.plot(xPoint, yPoint, "o") 

# Shows the graph
plt.show()
```

```python
xPoint = np.array([1,8,-2,-3,4])

# Used to create a plot plus a marker as {o}
plt.plot(xPoint, marker = "o") 
plt.show()
```

```python
xPoint = np.array([1,8,-1,3,4])

# Used to create a plot plus a marker as {*}
plt.plot(xPoint, marker = "*") 
plt.show()
```

```python
xPoint = np.array([1,8,3,4])

# Used to create a plot plus a marker as {o} and a color as {b}
plt.plot(xPoint,'o:r') 
plt.show()
```

```python
xPoint = np.array([1,8,6,2])

# Used to create a plot plus a marker as {o} and a size as {30}
plt.plot(xPoint, marker = "*", ms = 30)
plt.show()
```

```python
xPoint = np.array([2,3,6,1])

# Used to create a plot plus a marker as {o} and a size as {30} and with point background color as {red}
plt.plot(xPoint, marker = "*", ms = 30,mec = "r")
plt.show()
```

```python
xPoint = np.array([2,3,6,1])

# Used to create a plot plus a marker as {o} and a size as {30} and with point color as {red}
plt.plot(xPoint, marker = "*", ms = 30,mfc = "r")
plt.show()
```

```python
xPoint = np.array([2,3,6,1])

# plot with {dotted lines} and {o} marker ls - {dotted, dashed, dashdot}
plt.plot(xPoint, ls = "dashdot", marker = "o")
plt.show()
```

```python
xPoint = np.array([2,4,3,7])

# plot with {dotted lines} and {o} marker and {10} size of marker and {5} linewidth
plt.plot(xPoint, ls = "dotted", marker = "o",ms = 10, linewidth = 5)
plt.show()
```

```python
xPoint = np.array([1,3,6,2])
yPoint = np.array([2,2,7,1])
zPoint = np.array([1,9,3,0])

# Makes multiple line in a graphs with a title {Multiple Line}
plt.title("Multiple Line")
plt.plot(xPoint)
plt.plot(yPoint)
plt.plot(zPoint)
plt.show()
```

```python
xPoint = np.array([1,3,6,2])
yPoint = np.array([2,2,7,1])

# X-label and Y-label with a title
plt.title("Multiple Line")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.plot(xPoint,yPoint)
plt.show()  
```

```python
xPoint = np.array([1,4,2,9])
yPoint = np.array([4,9,3,2])

plt.title("Multiple Line")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
# Create a grid in the plot
plt.grid(True)
plt.plot(xPoint)
plt.plot(yPoint)
plt.show()
```

```python
xPoint = np.array([1,4,2,9])
yPoint = np.array([4,9,3,2])

# Subplot (row, column, index)
plt.subplot(1,2,1)
plt.plot(xPoint,yPoint)

x = np.array([1,4,5,3])
y = np.array([4,9,3,2])
plt.subplot(1,2,2)
plt.plot(x,y)
plt.show()
```

```python
lunch = np.array([1,3,2,7,9,1,2,3])
dinner = np.array([4,9,3,2,2,3,4,5])

# Scatter plot
plt.xlabel("Lunch")
plt.ylabel("Dinner")
plt.scatter(lunch,dinner) 
plt.show()
```

```python
x1 = np.array([13,4,5,6])
y1 = np.array([12,4,45,6])
# Giving Dynamic color
clr = np.array(["red", "green", "blue", "yellow"])
#Giving Dynamic Size
sz = np.array([100,210,340,410])
plt.scatter(x1, y1,c = clr,s = sz)
plt.show()
```

```python
x1 = np.array([0,0,0,0])
y1 = np.array([0,0,0,0])

# Giving Dynamic color
clr = np.array(["red", "green", "blue", "yellow"])

#Giving Dynamic Size
sz = np.array([100,210,340,410])
plt.scatter(x1, y1,c = clr,s = sz,alpha=0.3)
plt.show()
```

```python
x1 = np.array([1,3,2,7])
y1 = np.array([6,3,7,8])

# Giving Dynamic color
clr = np.array(["red", "green", "blue", "yellow"])

#Giving Dynamic Size
sz = np.array([100,210,340,410])

# Creating a Color Bar cmap{nipy_spectral, viridis}
plt.scatter(x1, y1,c = clr,s = sz,alpha=0.3,cmap="nipy_spectral")
plt.colorbar()
plt.show()
```

```python
x1 = np.array(['A','B','C','D'])
y1 = np.array([6,2,8,4])

# Used to create a bar graph
plt.bar(x1,y1)
plt.show()
```

```python
x1 = np.array(['A','B','C','D'])
y1 = np.array([6,2,8,4])

# Used to create a bar graph
plt.bar(x1,y1,color = "red")
plt.show()
```

```python
x1 = np.array(['A','B','C','D'])
y1 = np.array([6,2,8,4])

# Used to change the width of the graoh with edgecolor black
plt.bar(x1,y1,color = "pink",width=0.6, edgecolor = "black")
plt.show()
```

```python
x = np.random.randint(10,80,60)

# Used to create a histogram
plt.hist(x,color="lavender", edgecolor="black")
plt.show()
```

```python
x = np.array([6,2,8,4])

#Gives a name to every part
myLabels = ["Berries","Banana","Guava","Apple"]
clr = ["lavender","pink","violet","indigo"]

# Used to cut out a section
myExplode = [0,0.2,0,0]

# Used to create a pie chart with labels = {myLabels}, explodes the {myExplode}, autopct is used to gives the percentange
plt.pie(x, labels=myLabels, startangle=90,explode=myExplode,shadow=True, colors=clr,autopct='%1.0f%%')

# Shows the routing through every part of the chart
plt.legend(title = "Values")
plt.show()
```
